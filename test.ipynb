```python
from collections import deque

def bfs_dfs(graph, source):
    bfs, dfs = [], []
    
    if source in graph:
        visited_dfs = set()  # dfs
        def traverse(vertex):
            visited_dfs.add(vertex)
            dfs.append(vertex)
            for neighbour in graph[vertex]:
                if neighbour not in visited_dfs:
                    traverse(neighbour)
        traverse(source)

        visited_bfs = set()   # bfs
        queue = deque([source])
        while queue:
            vertex = queue.popleft()
            bfs.append(vertex)
            for neighbour in graph[vertex]:
                if neighbour not in visited_bfs:
                    visited_bfs.add(neighbour)
                    queue.append(neighbour)

        print("BFS:", bfs)
        print("DFS:", dfs)
    
    else:
        print("Invalid Source")



graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['F'],
    'D': ['E'],
    'E': ['F'],
    'F': []
}

bfs_dfs(graph, input())
```

```python
def uniform_cost_search(graph, start, goal):
    priority_queue = [(0 + huristic[start], start, [], 0)]
    visited = set()

    while priority_queue:
        tot_cost, current_node, path, cost_so_far = heapq.heappop(priority_queue)    # Popping the least cost
        if current_node == goal:
            return path + [current_node], cost_so_far, tot_cost
        if current_node in visited:
            continue
        visited.add(current_node)

        for neighbour, cost in graph[current_node].items():
            if neighbour not in visited:
                new_cost_so_far = cost_so_far + cost
                tot_cost = new_cost_so_far + huristic[neighbour]

                heapq.heappush(priority_queue, (tot_cost, neighbour, path + [current_node], new_cost_so_far))
        
    return None, float('inf')
```

```python
# Hill climb algorithm

def hil_climb_algo(numbers):
    current_index = 0
    while True:
        #check if next index is within the list range
        if current_index + 1 < len(numbers):
            #compare with the next number
            if numbers[current_index] < numbers[current_index + 1]:
                #update the neighbour
                current_index += 1
            else:
                #end of the list
                return numbers[current_index]

numbers = [1, 3, 5, 9]
max_number = hil_climb_algo(numbers)
max_number
```

```python
import random

def objective_function(x):
    # return -(x - 3)**4 + 9
    return 100 - (x-7)**2

def simple_hill_climbing(start, step_size = 1, epochs = 3):
    current = start
    curr_value = objective_function(current)

    for _ in range(epochs):
        # generate a neighbour state
        neighbour = current + random.choice([-step_size, step_size])
        neighbour_value = objective_function(neighbour)

        # if neighbour is better, move to it
        if neighbour_value > curr_value:
            current = neighbour
            curr_value = neighbour_value
            print(f"Simple HC: Moving to {current} with value {curr_value}")

    return current, curr_value

def steepest_ascent_hill_climbing(start, step_size = 1, epochs = 3):
    current = start
    curr_value = objective_function(current)

    for _ in range(epochs):
        best_neighbour = None
        best_neighbour_value = float('-inf')

        for neighbour in [current + step_size, current - step_size]:
            neighbour_value = objective_function(neighbour)
            if neighbour_value > best_neighbour_value:
                best_neighbour = neighbour
                best_neighbour_value = neighbour_value

        if best_neighbour_value > curr_value:
            current = best_neighbour
            curr_value = best_neighbour_value
            print(f"Steepest HC: Moving to {current} with value {curr_value}")

    return current, curr_value

simple_hill_climbing(1, step_size = 1, epochs = 100)
```

```python
class NaiveBayes:
    def __init__(self):
        self.probabilities = {}
        self.class_probabilities = {}

    def fit(self, X, y):
        self.data = pd.DataFrame(X)
        self.data['target'] = y
        self.class_probabilities = self.data['target'].value_counts(normalize = True).to_dict()

        for column in self.data.columns:
            if column != 'target':
                self.probabilities[column] = self.data.groupby(['target', column]).size().unstack().apply(lambda x: x / x.sum(), axis = 1).to_dict()

    def predict(self, X):
        predictions = []
        for i in range(len(X)):
            prediction = {}
            for target in self.class_probabilities:
                prob = self.class_probabilities[target]
                for column in range(len(X[i])):
                    if column != 'target':
                        prob *= self.probabilities[column].get((target, X[i][column]), 0)
                prediction[target] = prob
            predictions.append(max(prediction, key = prediction.get))
        return predictions

    def score(self, X, y):
        predictions = self.predict(X)
        return sum(predictions == y) / len(y)
```

# Memoization

```python
import time

cache = {0: 0, 1: 1}
total_calls = 0

# Using cache system to reduce number of calls in recursion
def fibo_recursive(n):
    global total_calls, cache

    total_calls += 1

    if n in cache: 
        return cache[n]
    
    result = fibo_recursive(n-1) + fibo_recursive(n-2)
    cache[n] = result
    
    return result

# Fibonacci using Recursion
n = int(input())
start = time.time()
for _ in range(n):
    print(f'{fibo_recursive(_)}', end = ' ')
end = time.time()
print(f'\nTotal Time: {(end - start)*1000}s')
print(f'Total Calls: {total_calls}')
```
