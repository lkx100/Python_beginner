```python
product_data = df[['Category', 'Sales', 'Quantity', 'Discount', 'Profit']]
```

```python
product_data.sample(10)
```

```python
plt.bar(product_data['Category'], product_data['Sales'])
plt.xlabel("Category")
plt.ylabel("Sales")
```

```python
plt.bar(product_data['Category'], product_data['Profit'])
plt.xlabel("Category")
plt.ylabel("Profit")
```
